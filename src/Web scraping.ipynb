```python
url = "https://www.gamedesigning.org/game-development-studios/"
```

```python
res= requests.get(url)
html=res.text
html[:10]
soup = BeautifulSoup(html, "html.parser")
games = soup.find_all('h2')
```

```python
company=[]
for i in games: 
    company.append(i.text)
```

```python
listina=[]
for e in company: 
    listina.append(re.findall(r"^[0-9]+\.\s.+", e))   
```

```python
li=[]
for i in listina:
    if not i == []:
        li.append(i)
```

```python
enterprise=[]
for i in li:
    for e in i:
        enterprise.append(e)
```

```python
res= requests.get(url)
html=res.text
html[:10]
soup = BeautifulSoup(html, "html.parser")
gamis = soup.find_all('p')
```

```python
lissta=[]
for i in gamis:
    lissta.append(i.text.strip())
```

```python
listin=[]
for e in lissta: 
    listin.append(re.findall(r"^Headquarters.+", e))   
```

```python
l=[]
for i in listin:
    if not i == []:
        l.append(i)
print(l)
```

```python
localitation=[]
for i in l:
    for e in i:
        localitation.append(e)
```

```python
dic={"Big50":enterprise,"country":localitation}
df=pd.DataFrame(dic)
```

```python
def clean(data):
    data['Big50'] = data['Big50'].str.replace("\d+\.\s", '')
    data['country'] = data['country'].str.replace("Headquarters:\s", '')
    return data
```

```python
to_sust= {
    "(.+)?Japan(.+)?":"Japan",
    "(.+)?USA(.+)?":"USA",
    "(.+)?Canada(.+)?":"Canada",
    "(.+)?US(.+)?":"USA",
    "(.+)?UK(.+)?":"UK",
    "(.+)?California(.+)?":"USA",
    "(.+)?NY(.+)?":"USA",
    "(.+)?WA(.+)?":"USA",
    "(.+)?City(.+)?":"USA",
    "(.+)?Cary(.+)?":"USA",
    "(.+)?Sweden(.+)?":"Sweden",
    "(.+)?France(.+)?":"France",
    "(.+)?Korea(.+)?":"Korea",
    "(.+)?Russia(.+)?":"Russia",
}
```

```python
def getcountry(data, lista):
    for k,v in lista.items():
        data = data.str.replace(k,v)
    return data
```

```python
new_column=getcountry(df.country,to_sust)
```

```python
df["countries"]=new_column
df=df[["Big50","countries"]]
```
