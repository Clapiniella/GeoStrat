```python
import pandas as pd
```

```python
dist=pd.read_csv('distoair.csv')
```

```python
dist = dist.sort_values('dis to an airport',ascending=True)
```

```python
cutoffs = [0,10,20,50,100,300]
km_labels = ['0-10km', '10-20km', '20-50km','50-100km','100-300km']
dist['range'] = pd.cut(dist['dis to an airport'], cutoffs, labels = km_labels)
dist['range'].value_counts()
```
