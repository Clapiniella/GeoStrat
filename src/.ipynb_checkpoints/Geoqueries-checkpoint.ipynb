```python
import folium
import pandas as pd
from pymongo import MongoClient
import json
```

```python
def connectCollection(database, collection):
    client = MongoClient()
    db = client[database]
    coll = db[collection]
    return db, coll
```

```python
#col--> db[collection]
def geoquery(geoindex,distance,col):
    preferred= col.find(
        {"location": 
         {"$near": 
          {"$geometry":
           {geoindex
           }, 
           "$maxDistance":distance
            }
        }
    })
    return list(preferred)
```
