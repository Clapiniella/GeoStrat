# Map & Geoqueries

```python
import folium
import pandas as pd
from pymongo import MongoClient
import json
```

```python
def connectCollection(database, collection):
    client = MongoClient()
    db = client[database]
    coll = db[collection]
    return db, coll
```

```python
db1, coll1 = connectCollection('companies','preferred')
```

```python
preferred= coll1.find(
    {"location": 
     {"$near": 
      {"$geometry":
       {"type": "Point",
        "coordinates":[-122.4194155, 37.7749295]
       }, 
       "$maxDistance": 10000
        }
    }
})
preferred=list(preferred)
```

```python
db, coll2 = connectCollection('companies','starbucks')
```

```python
starbucks= coll2.find(
    {"location": 
     {"$near": 
      {"$geometry":
       {"type": "Point",
        "coordinates":[-122.4194155, 37.7749295]
       }, 
       "$maxDistance": 10000
        }
    }
})
starbucks=list(starbucks)
```

```python
db, coll3 = connectCollection('companies','schools')
```

```python
schools= coll3.find(
    {"location": 
     {"$near": 
      {"$geometry":
       {"type": "Point",
        "coordinates":[-122.4194155, 37.7749295]
       }, 
       "$maxDistance": 10000
        }
    }
})
schools=list(schools)
```

```python
db, coll4 = connectCollection('companies','kindergarten')
```

```python
kindergartens= coll4.find(
    {"location": 
     {"$near": 
      {"$geometry":
       {"type": "Point",
        "coordinates":[-122.4194155, 37.7749295]
       }, 
       "$maxDistance": 10000
        }
    }
})
kindergartens=list(kindergartens)
```

```python
db, coll5= connectCollection('companies','nightclub')
```

```python
nightclubs= coll5.find(
    {"location": 
     {"$near": 
      {"$geometry":
       {"type": "Point",
        "coordinates":[-122.4194155, 37.7749295]
       }, 
       "$maxDistance": 10000
        }
    }
})
nightclubs=list(nightclubs)
```

```python
nightclubs
```

```python
map_city=folium.Map(location=[37.7749295,-122.4194155], zoom_start=12)
for star in starbucks:
    folium.Marker(star['location']['coordinates'][::-1],
                    radius=2,
                    icon=folium.Icon(icon='glyphicon glyphicon-cutlery',color='green', popup=star['name']), 
                   ).add_to(map_city)
    
for school in schools:
    folium.Marker(school['location']['coordinates'][::-1],
                    radius=2,
                    icon=folium.Icon(icon='glyphicon glyphicon-blackboard',color='pink'), 
                   ).add_to(map_city)
    
for kinder in kindergartens:
    folium.Marker(kinder['location']['coordinates'][::-1],
                    radius=2,
                    icon=folium.Icon(icon='glyphicon glyphicon-baby-formula',color='blue'), 
                   ).add_to(map_city)
    
for night in nightclubs:
    folium.Marker(night['location']['coordinates'][::-1],
                    radius=2,
                    icon=folium.Icon(icon='glyphicon glyphicon-glass',color='red'), 
                   ).add_to(map_city)
for pre in preferred:
    folium.Marker(pre['location']['coordinates'][::-1],
                    radius=2,
                    icon=folium.Icon(icon='glyphicon glyphicon-piggy-bank',color='black', popup=star['name']), 
                   ).add_to(map_city)
    
```
