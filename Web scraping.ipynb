```python
url = "https://www.gamedesigning.org/game-development-studios/"
```

```python
res= requests.get(url)
html=res.text
html[:10]
soup = BeautifulSoup(html, "html.parser")
games = soup.find_all('h2')
```

```python
company=[]
for i in games: 
    company.append(i.text)
```

```python
print(company)
```

```python
listina=[]
for e in company: 
    listina.append(re.findall(r"^[0-9]+\.\s.+", e))   
```

```python
listina
```

```python
li=[]
for i in listina:
    if not i == []:
        li.append(i)
print(li)
```

```python
enterprise=[]
for i in li:
    for e in i:
        enterprise.append(e)
```

```python
res= requests.get(url)
html=res.text
html[:10]
soup = BeautifulSoup(html, "html.parser")
gamis = soup.find_all('p')
gamis
```

```python
lissta=[]
for i in gamis:
    lissta.append(i.text.strip())
```

```python
lissta
```

```python
listin=[]
for e in lissta: 
    listin.append(re.findall(r"^Headquarters.+", e))   
```

```python
listin
```

```python
l=[]
for i in listin:
    if not i == []:
        l.append(i)
print(l)
```

```python
print(len(l))
```

```python
localitation=[]
for i in l:
    for e in i:
        localitation.append(e)
```

```python
localitation
```

```python
dic={"Big50":enterprise,"name":localitation}
df=pd.DataFrame(dic)
```

```python
def clean(data):
    data['Big50'] = data['Big50'].str.replace("\d+\.\s", '')
    data['name'] = data['name'].str.replace("Headquarters:\s", '')
    return data
```
